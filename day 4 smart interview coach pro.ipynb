```python
# Smart Interview Coach Pro

## Project Overview
This Python application simulates a technical interview, evaluates answers, tracks scores, and generates a performance report.

### Features
- Interactive interview questions
- Score tracking
- Percentage calculation
- Performance assessment
- Interview readiness report
```

```python
print("=" * 50)
print("        SMART INTERVIEW COACH PRO")
print("=" * 50)

score = 0

# Question 1
print("\nQuestion 1")
print("Which keyword is used to create a function in Python?")

answer = input("Your Answer: ")

if answer.lower() == "def":
    print("✓ Correct")
    score += 1
else:
    print("✗ Wrong")
    print("Correct Answer: def")

# Question 2
print("\nQuestion 2")
print("Which data type stores multiple items?")
print("Hint: list, int, float, string")

answer = input("Your Answer: ")

if answer.lower() == "list":
    print("✓ Correct")
    score += 1
else:
    print("✗ Wrong")
    print("Correct Answer: list")

# Question 3
print("\nQuestion 3")
print("Which loop is used when the number of iterations is known?")
print("Hint: for or while")

answer = input("Your Answer: ")

if answer.lower() == "for":
    print("✓ Correct")
    score += 1
else:
    print("✗ Wrong")
    print("Correct Answer: for")

# Question 4
print("\nQuestion 4")
print("What function is used to get user input?")

answer = input("Your Answer: ")

if answer.lower() == "input":
    print("✓ Correct")
    score += 1
else:
    print("✗ Wrong")
    print("Correct Answer: input")

# Question 5
print("\nQuestion 5")
print("What function displays output on the screen?")

answer = input("Your Answer: ")

if answer.lower() == "print":
    print("✓ Correct")
    score += 1
else:
    print("✗ Wrong")
    print("Correct Answer: print")

# Final Report
print("\n")
print("=" * 50)
print("           INTERVIEW REPORT")
print("=" * 50)

print("Total Score:", score, "/ 5")

percentage = (score / 5) * 100

print("Percentage:", percentage, "%")

if percentage >= 80:
    print("Result: Excellent!")
    print("You are interview ready.")
elif percentage >= 60:
    print("Result: Good")
    print("Need some more practice.")
elif percentage >= 40:
    print("Result: Average")
    print("Keep learning Python.")
else:
    print("Result: Beginner")
    print("Practice daily and try again.")

print("\nThank you for using Smart Interview Coach Pro")
```
